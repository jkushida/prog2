# 動的なメモリ確保

## 

以下は`malloc`関数で, **int型の要素を格納するメモリ領域を動的に確保する**プログラムである．

実行結果と同様の出力が表示されるように, プログラムを完成させよ．

```c
#include <stdio.h>
#include <stdlib.h>

int main(void)
{
    int *ip;        // int型のポインタipを宣言
    int i, n;
    
    n = 5;
    ip = (int *)malloc(n * sizeof(int));  // メモリ領域の確保
    
    for (i=0; i<n; i++){
        /* コードを書く */
    }
    
    for (i=0; i<n; i++)
        printf("%d ", ip[i]);
    printf("\n");
    
    free(ip);       // 確保したメモリ領域の解放
    
    return 0;
}
```

実行結果
```
5 4 3 2 1 
```

## 

以下は`malloc`関数で, char型の要素を格納するメモリ領域を動的に確保するプログラムである．

実行結果と同様の出力が表示されるように, プログラムを完成させよ．

```c
#include <stdio.h>
#include <stdlib.h>

int main(void)
{
    char *cp;        // char型のポインタcpを宣言
    int i, n;
    
    // メモリ領域の確保
    /* コードを書く */
     
    // 文字の代入
    for (i=0; i<n-1; i++){
        *(cp+i)='a' + i;
    }
    *(cp+i)='\0';
    
    printf("%s\n", cp);
    
    free(cp);       // 確保したメモリ領域の解放
    
    return 0;
}
```
