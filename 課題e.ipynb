# ファイルと構造体

## 

次の処理を行うプログラムを作成せよ．

1. 作業フォルダに **seiseki.dat** というファイルを作成する
2. キーボードから受験番号, 英語, 数学, 国語の得点を入力する．
2. 3科目の合計点を計算する．
3. 受験番号, 英語, 数学, 国語の得点, 合計点をファイルに書き込む．
4. 受験番号が0以下のときは, 処理を終了する．


実行例

- 端末での入力

```
受験番号 => 101
英語　数学　国語の点 => 70 80 90
受験番号 => 102
英語　数学　国語の点 => 80 90 100
受験番号 => 103
英語　数学　国語の点 => 60 70 60
受験番号 => 0

```

- 作成された seiseki.dat の中身

```
101 70 80 90 240
102 80 90 100 270
103 60 70 60 190
```


````c
#include <stdio.h>

int main(void)
{
    int    jno;
    int    ei, suu, kok, total;
    FILE   *outfp;
    
    outfp = fopen("seiseki.dat","w");
    printf("受験番号 => ");
    scanf("%d",&jno);
    while(jno>0) {
        printf("英語　数学　国語の点 => ");
        scanf("%d %d %d", &ei, &suu, &kok);
        total = ei + suu + kok;
        
        fprintf(outfp,"%d %d %d %d %d\n", jno, ei, suu, kok, total);
        printf("受験番号 => ");
        scanf("%d", &jno);
    }
    fclose(outfp);
    
    return 0;
}
````

## 

問題xx（前の問題）で作成した **seiseki.dat** を開き, ファイルから受験番号と英語, 数学, 国語の得点を読み出し，以下のように出力するプログラムを作成せよ．

実行例

```
受験番号　英語　数学　国語　合計
    101    70    80    90   240
    102    80    90   100   270
    103    60    70    60   190
```

````c
/*  例題８−４  */
#include <stdio.h>

int  main(void)
{
    int    jno;
    int    ei, suu, kok, kei;
    FILE   *infp;
    infp = fopen("seiseki.dat","r");
    printf("受験番号　英語　数学　国語　合計\n");
    while(fscanf(infp,"%d %d %d %d %d", &jno, &ei, &suu, &kok, &kei) != EOF)
    {
        kei = ei + suu + kok;
        printf("%7d %5d %5d %5d %5d\n", jno, ei, suu, kok, kei);
    }
    fclose(infp);
    return 0;
}
````

## 

以下の表の３人分の番号，氏名，年齢，勤務年数のデータを**構造体配列**に格納せよ．

また，以下の実行結果のように表示せよ．

ただし, 
- 構造体のタグ名は member とすること.
```c
struct member
```
- for文を用いること.

|番号 |	 氏名 	| 年齢 	|勤務年数|
| :---:  | :---:  | :---: | :---:  |
|1 	  |  田中 	|     21 	   | 4 |
|2 	  |  山田 	 |    19 	    | 3 |
|3 	  |  坂本 	 |    20 	    | 5 |


実行結果
```
番号 	 氏名 	 年齢 	勤務年数
1 	    田中 	     21 	     4 
2 	    山田 	     19 	     3 
3 	    坂本 	     20 	     5 
```

````c
/* 演習問題８−２ */
#include <stdio.h>
struct  member
{
    int     no;
    char    name[10];
    int     age;
    int     exp;
};

int main(void)
{
    struct  member s[3]={
        {1, "田中", 21, 4},
        {2, "山田", 19, 3},
        {3, "坂本", 20, 5}
    };
    int  i;
    
    printf ("番号 \t 氏名 \t 年齢 \t勤務年数\n");
    for (i=0; i<3; i++) {
        printf ("%d \t %s \t %d \t %d \n",s[i].no, s[i].name, s[i].age, s[i].exp);
    }
    return 0;
}
````

## 
前の問題xx で作成したプログラムにおいて, 構造体のポインタ変数を使ってメンバを参照し，出力するように変更せよ．

なお, 構造体変数へのポインタは以下のように宣言すること.
```c
struct  member *p;
```

出力結果は問題xxと同じである．




````c
/* 演習問題８−３ */
#include <stdio.h>

struct member
{
    int  no;
    char    name[10];
    int     age;
    int     exp;
};

int main(void)
{
    struct  member s[3]={
        {1, "田中", 21, 4},
        {2, "山田", 19, 3},
        {3, "坂本", 20, 5}
    };
    struct  member *p;
    int  i;
    
    printf ("番号 \t 氏名 \t 年齢 \t勤務年数\n");
    for (i=0; i<3; i++) {
        p = &s[i];
        printf ("%d \t %s \t %d \t %d \n",p->no, p->name, p->age, p->exp);
    }
    return 0;
}
````

## 

問題xxで作成したプログラムにおいて, 表示する部分をユーザ関数とするように変更せよ．

なお, ユーザ関数は以下のように定義すること
```c

```

````c
/* 演習問題８−４ */
#include <stdio.h>

struct  member
    {
    int     no;
    char    name[20];
    int     age;
    int     exp;
    };
void disp(struct member[]);

int main(void)
{
    struct  member s[3]={
        {1, "田中", 21, 4},
        {2, "山田", 19, 3},
        {3, "坂本", 20, 5}
    };
    disp(s);
    return 0;
}

void disp(struct member p[])
{
    int  i;
    
    printf ("番号 \t 氏名 \t 年齢 \t勤務年数\n");
    for (i=0; i<3; i++) {
        printf ("%d \t %s \t %d \t %d \n",p[i].no, p[i].name, p[i].age, p[i].exp);
    }
}
````

````c
// 解答
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

// 関数形式マクロの定義
// dice()は1から６の範囲の整数乱数を返す
#define dice() rand() % 6 + 1;

int main(void)
{
    FILE *fp;
    int i, p, x[7]={0};
    char *fname = "dice1.csv";
    char name[7][10];
    
    // 乱数シードを初期化
    srand(time(NULL));
    
    fp = fopen( fname, "w" );
    if( fp == NULL ){
        printf( "%sファイルが開けません\n", fname );
        return -1;
    }
    
    // ラベルの文字列を設定
    for (i=1; i<7; i++)
        sprintf(name[i], "x[%d]", i);
    
    for (i=0; i<6000; i++) {
        p = dice();
        
        x[p]++;      // サイコロ目の数をカウント
    }
    
    for (i=1; i<7; i++)
        fprintf( fp, "%s, %d\n", name[i], x[i]);
    
    fclose(fp);
    
    printf( "%sへのファイル書き込み終了\n", fname);

    return 0;
}
````
