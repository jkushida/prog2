以下のプログラムは，サイコロを6000回振ったとき，1~6の目が出る回数をカウントし端末に表示している．

このプログラムを変更し, 出力をcsvファイルに書き込むようにせよ．

なお, csvファイルのファイル名は, dice1.csv とすること．

```c
char *fname = "dice1.csv";
```


また， このプログラムではint型の配列 x は x[1] から x[6] までを使用している．

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

// 関数形式マクロの定義
// dice()は1から６の範囲の整数乱数を返す
#define dice() rand() % 6 + 1;

int main(void)
{
    int i, p, x[7]={0};
    char name[7][10];
    
    // 乱数シードを初期化
    srand(time(NULL));
    
    // ラベルの文字列を設定
    for (i=1; i<7; i++)
        sprintf(name[i], "x[%d]", i);
    
    for (i=0; i<6000; i++) {
        p = dice();
        
        x[p]++;      // サイコロ目の数をカウント
    }
    
    for (i=1; i<7; i++)
        printf("%s, %d\n", name[i], x[i]);

    return 0;
}
```

実行結果(端末への出力)

```
x[1], 966
x[2], 1017
x[3], 988
x[4], 1060
x[5], 971
x[6], 998
```

## 

以下は，区間$[𝑎,𝑏]$で連続な関数$𝑦=𝑓(𝑥)$の定積分を行うプログラムである．

課題eの説明のスライドを参考にし，プログラムを完成させよ．


```c
// サンプルコード
#include <stdlib.h>
#include <stdio.h>
#include <math.h>

// 対象の関数
double func(double x);

// 長方形近似による区分求積法
double rectangle(double a, double b, int n);

int main(void)
{
    int n;
    double a, b;
    double r1;
    
    // 区間[a,b]の面積を求める
    a = 0;
    b = 1;
    
    printf("n, rectangle\n");
    for(n=10; n<=100; n+=10){
        r1=rectangle(a, b, n);
        printf("%d, %f\n", n, r1);
    }
    
    return 0;
}

double func(double x){
    return(1 / (1 + x*x));
}

double rectangle(double a, double b, int n){
    double x, dx, h, sum, ds;
    
    dx = (b-a) / n;
    sum = 0;
    x = a;
    
    /* 面積の計算 */
    while (x < b) {
        h = func(x);
        ds = h * dx;
        sum += ds;
        x += dx;
    }
    return sum;
}
```
