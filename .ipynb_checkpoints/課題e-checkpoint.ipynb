# 構造体の基本

## 

以下は, **int型の no,  char型の配列 name[10],   int型の point をメンバとして持つ構造体**を用いたプログラムである．

実行結果と同様の出力が表示されるように, プログラムを完成させよ．

```c
#include <stdio.h>

/* 構造体の枠型の宣言を書く */

int main()
{
    struct student seito1 = { 1001, "Tanaka", 70};
    
    printf( "%d %s %d\n", seito1.no, seito1.name, seito1.point);
    return 0;
}
```

実行結果
```
1001 Tanaka 70
```

````c
// 解答
#include <stdio.h>

struct student{
    int no;
    char name[10];
    int point;
};

int main()
{
    struct student seito1 = { 1001, "Tanaka", 70};
    
    printf( "%d %s %d\n", seito1.no, seito1.name, seito1.point);
    return 0;
}
````

## 

以下は, 構造体に格納された学生の情報を表示するプログラムである．

**再試験を行った結果，情報が更新されている．**


実行結果と同様の出力が表示されるように, プログラムを完成させよ．

```c
#include <stdio.h>

// 構造体の枠型の宣言
struct student{
    int no;           // 学籍番号
    char name[10];    // 氏名
    int test1;        // テスト1の点
    int test2;        // テスト2の点
    double average;   // テストの平均点
};

int main()
{
    struct student seito1 = {5, "Yamada", 80, 70};
    
    /* コードを書く */
    
    // 構造体の持つ情報を表示
    printf("更新前 学生番号: %d,  氏名: %s, 平均点: %f\n", seito1.no, seito1.name, seito1.average);
    
    // 再試験
    seito1.test1 = 85;
    seito1.test2 = 75;
    
    /* コードを書く */
    //　更新後の情報を表示
    
    return 0;
}
```

実行結果
```
更新前 学生番号: 5,  氏名: Yamada, 平均点: 75.0
更新後 学生番号: 5,  氏名: Yamada, 平均点: 80.0

```

````c
// 解答
#include <stdio.h>

// 構造体の枠型の宣言
struct student{
    int no;           // 学籍番号
    char name[10];    // 氏名
    int test1;        // テスト1の点
    int test2;        // テスト2の点
    double average;   // テストの平均点
};

int main()
{
    struct student seito1 = {5, "Yamada", 80, 70};
    
    seito1.average = (seito1.test1 + seito1.test2)/2.0;
    
    // 構造体の持つ情報を表示
    printf("更新前 学生番号: %d,  氏名: %s, 平均点: %0.1f\n", seito1.no, seito1.name, seito1.average);
    
    // 再試験
    seito1.test1 = 85;
    seito1.test2 = 75;
    
    seito1.average = (seito1.test1 + seito1.test2)/2.0;
    
    printf("更新後 学生番号: %d,  氏名: %s, 平均点: %0.1f\n", seito1.no, seito1.name, seito1.average);
    
    return 0;
}
````

## 

以下は, 構造体srcの情報を同じ型の構造体であるdestにコピーするプログラムである．


実行結果と同様の出力が表示されるように, プログラムを完成させよ．

hint: 構造体のコピー 同じ型


```c
#include <stdio.h>

// 構造体の枠型の宣言
struct INT_MEMBERS {
    int a, b, c;
};

int main(void)
{
    struct INT_MEMBERS src, dest;
    
    src.a = 1;
    src.b = 2;
    src.c = 3;
    
    /* コードを書く */
    
    printf("dest.a = %d\n", dest.a);
    printf("dest.b = %d\n", dest.b);
    printf("dest.b = %d\n", dest.c);
    
    return 0;
}

```

実行結果
```
dest.a = 1
dest.b = 2
dest.b = 3
```

````c
// 解答
#include <stdio.h>

// 構造体の枠型の宣言
struct INT_MEMBERS {
    int a, b, c;
};

int main(void)
{
    struct INT_MEMBERS src, dest;

    src.a = 1;
    src.b = 2;
    src.c = 3;

    // 構造体のコピー
    dest = src;

    printf("dest.a = %d\n", dest.a);
    printf("dest.b = %d\n", dest.b);
    printf("dest.b = %d\n", dest.c);

    return 0;
}
````

```c
typedef struct{
```

```c
#include <stdio.h>

// 構造体の定義
struct student{
    int no;
    char name[5];
    double average;
};

{
    int i;
    // 構造体配列の宣言と初期化
    struct student seito[3] = {

        {2, "Kei", 56.8},
        {3, "Taro", 65.2},
    };

    for(i = 0; i<3; i++)
        printf("学生番号: %d,  氏名: %s,  平均点: %.1f\n",seito[i].no, seito[i].name, seito[i].average);
```

````c
// 解答
#include <stdio.h>

// 構造体の定義
typedef struct{
    int no;
    char name[5];
    double average;
}seito_t;

int main()
{
    int i;
    // 構造体配列の宣言と初期化
    seito_t seito[3] = {
        {1, "Taka", 65.1},
        {2, "Kei", 56.8},
        {3, "Taro", 65.2},
    };

    for(i = 0; i<3; i++)
        printf("学生番号: %d,  氏名: %s,  平均点: %.1f\n",seito[i].no, seito[i].name, seito[i].average);

    return (0);
}
````

以下のプログラムでは，構造体に５件の社員のデータを格納している．

```c
#include <stdio.h>

typedef struct{ // 社員情報

}syain_dt;

void print_one_data(syain_dt *sp);

        {103, "Sasaki", 450000},
        {104, "Yamada", 250000},

    };
    syain_dt *sp; //構造体のポインタの宣言
    
    // コードを書く

    return 0;
}

// 一人分のデータを表示する関数

    printf("%d \t %s \t %.1f\n",sp->no,  sp->name,  sp->kyuuyo);
```

````c
#include <stdio.h>

typedef struct{ // 社員情報
    int no;
    char name[8];
    double kyuuyo;
}syain_dt;

void print_one_data(syain_dt *sp);

int main(){
    int i;
    syain_dt ics[5] = {
        {101, "Sakata", 300000},
        {102, "Tanaka", 500000},
        {103, "Sasaki", 450000},
        {104, "Yamada", 250000},
        {105, "harada", 370000},
    };
    syain_dt *sp; //構造体のポインタの宣言
    
    sp = ics;
    for(i=0; i<5; i++)
        print_one_data(sp+i);

    return 0;
}

// 一人分のデータを表示する関数
void print_one_data(syain_dt *sp){
    printf("%d \t %s \t %.1f\n",sp->no,  sp->name,  sp->kyuuyo);
}
````

2. averageの値により, hanteiを決定．
    - averageが60未満なら，hanteiに"不合格"を代入
    - averageが60点以上なら，hanteiに"合格"を代入

```c
    student_dt info[N] = {
        {101, "Sakata", {90, 60, 65}, 0, "-"},
        {102, "Tanaka", {56, 65, 78}, 0, "-"},
        {103, "Sasaki", {92, 89, 58}, 0, "-"},
        {104, "Yamada", {35, 45, 67}, 0, "-"},
        {105, "harada", {23, 87, 36}, 0, "-"},
    };
    printf("成績処理前\n");
    for(i=0; i<N; i++)
        print_one_data(info[i]);
    
    // 配列score[3]の要素の平均からhanteiを決定する.

        seiseki_syori(&info[i]);
    
    printf("成績処理後\n");
    for(i=0; i<N; i++)
        print_one_data(info[i]);
    return 0;
}

    printf("%.1f ", sp.average);

        strcpy(sp->hantei, "合格");
```

````c
#include <stdio.h>

typedef struct{ // 社員情報
    int no;
    char name[8];
    double kyuuyo;
}syain_dt;

void print_one_data(syain_dt sp);
void swap(syain_dt *d1, syain_dt *d2);

#define N 5

int main(){
    int i, j;
    syain_dt ics[N] = {
        {101, "Sakata", 300000},
        {102, "Tanaka", 500000},
        {103, "Sasaki", 450000},
        {104, "Yamada", 250000},
        {105, "harada", 370000},
    };
    printf("ソート前\n");
    for(i=0; i<N; i++)
        print_one_data(ics[i]);
    
    //BubbleSort
    // kyuuyoをキーとして，構造体icsを昇順にソート
    for(i=0; i<N-1; i++){
        for(j=N-1; j>i; j--){
            if(ics[j-1].kyuuyo > ics[j].kyuuyo){
                swap( &ics[j], &ics[j-1] );
            }
        }
    }
    
    printf("ソート後\n");
    for(i=0; i<N; i++)
        print_one_data(ics[i]);

    return 0;
}

// 一人分のデータを表示する関数
void print_one_data(syain_dt sp){
    printf("%d \t %s \t %.1f\n",sp.no,  sp.name,  sp.kyuuyo);
}

void swap(syain_dt *d1, syain_dt *d2){
    syain_dt tmp;
    tmp = *d1;
    *d1 = *d2;
    *d2 = tmp;
}
````
