以下は配列を用いた**キュー**のプログラムである．

```c
#include <stdio.h>
#include <string.h>

/* キューに入るデータの上限 */
#define MAX_QUEUE_NUM 10

/* キュー構造体 */
typedef struct{
  /* エンキューされているデータの数 */
  int num;
  /* キューを配列dataで表現*/
  int data[MAX_QUEUE_NUM];
} QUEUE_T;

void initQueue(QUEUE_T *queue);
void printQueue(QUEUE_T *queue);
void enqueue( /* コードを書く */ );
int dequeue( /* コードを書く */ );

int main(void){

  int m;
  int input;
  int output;
  QUEUE_T queue;

  /* キューを初期化 */
  initQueue(&queue);

  while(1){
      
    printf("\n----------\n");
    printf("メニュー\n 1:エンキュー\n 2:デキュー\n 3:キュー表示\n それ以外:終了\n");
    printf("----------\n");

    printf("input > ");
    scanf("%d", &m);

    /* 選ばれたメニュー (m) に応じて処理を分岐 */
    if(m == 1){
      printf("エンキューする数は？（0以上の整数のみ可）> ");
      scanf("%d", &input);
      if(input < 0){
        printf("負の数は受け付けていません\n");
        continue;
      }
      printf("%dをエンキューします\n", input);
      enqueue(&queue, input);
    }else if(m == 2){
      printf("デキューします\n");
      output = dequeue(&queue);
      if(output != -1){
        printf("%dをデキューしました\n", output);
      }
    } else if(m == 3){
      printf("キューの中身を表示します\n");
      printQueue(&queue);
    } else {
      /* 1, 2, 3以外の場合はwhile文を抜け出すためにbreakを実行 */
      printf("終了します\n");
      break;
    }
  }
  return 0;
}

/* キューを初期化する関数 */
void initQueue(QUEUE_T *queue){
  int i = 0;
  queue->num = 0;

  for(i = 0; i < MAX_QUEUE_NUM; i++){
    queue->data[i] = 0;
  }
}

/* キューの中身を表示 */
void printQueue(QUEUE_T *queue){
  int i = 0;
    
  printf("(左側がキューの先頭)\n");
  for(i = 0; i < queue->num; i++){
    printf("%d,", queue->data[i]);
  }
  printf("\n");
}

/* エンキューする関数 */
void enqueue( /* コードを書く */ ){

  /* キューが満杯なら何もせず関数終了 */
  if(queue->num >= MAX_QUEUE_NUM){
    printf("キューが満杯でエンキューできません\n");
    return;
  }
    
  // コードを書く

}

/* デキューする関数 */
int dequeue( /* コードを書く */ ){
  int i = 0;
  int ret = 0;

  /* キューが空なら何もせずに関数終了 */
  if(queue->num == 0){
    printf("キューが空です\n");
    return -1;
  }
    
  //コードを書く
}


```

```c
// f5-1.cのコード
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct{
    int no;
    char name[5];
    int score1;
    int score2;
    double point;
}student_t;

void print_data(student_t *seito, int n);

// 関数のプロトタイプ宣言を追加

int main()
{
    int mode;
    int n = 5; // データ数
    // 構造体配列の宣言と初期化
    student_t seito[5]={
        {1001, "Taka", 75, 60, 65.1},
        {1002, "Miki", 81, 34, 56.8},
        {1003, "Taro", 79, 56, 67.9},
        {1004, "Jiro", 80, 67, 87.4},
        {1005, "Yuki", 93, 89, 95.2}
        };
    
    // ソート前のデータを表示
    print_data(seito, n);
    
    printf("select mode [1, 2, 3] ->  ");
    scanf("%d", &mode);
    
    // mode によりソートの方法を選択
    switch (mode){
        case 1:
            printf("score1をキーとして昇順にソート\n\n");
            // 関数を呼び出す
            break;
            
        case 2:
            printf("score2をキーとして降順にソート\n\n");
            // 関数を呼び出す
            break;
            
        case 3:
            printf("pointをキーとして降順にソート\n\n");
            // 関数を呼び出す
            break;
            
        default:
            printf("error -> input [1, 2, 3]\n");
            exit (1);
    }
    
    // ソート後のデータを表示
    print_data(seito, n);
    
    return 0;
}

// 構造体配列を表示する関数
void print_data(student_t *seito, int n){
    int i;
    
    printf("no, name, score1, score2, point\n");
    for(i = 0; i<n; i++)
        printf("%d, %s, %d, %d, %.1f\n",seito[i].no, seito[i].name, seito[i].score1, seito[i].score2, seito[i].point);
    printf("\n");
    
}

// 構造体用のswap関数
void swap(student_t *a, student_t *b){
    student_t temp;
    
    temp = *a;
    *a = *b;
    *b = temp;
    
}

// score1をキーとして昇順にソートする関数を作成


// score2をキーとして降順にソートする関数を作成


// pointをキーとして降順にソートする関数を作成


```
