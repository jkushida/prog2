# ファイルと構造体

## 

次の処理を行うプログラムを作成せよ．

1. キーボードから受験番号, 英語, 数学, 国語の得点を入力する．
2. 3科目の合計点を計算する．
3. 受験番号, 英語, 数学, 国語の得点, 合計点をファイル (**seiseki.dat**) に書き込む．
4. 受験番号が0以下のときは, 処理を終了する．


実行例

- 端末での入力

```
受験番号 => 101
英語　数学　国語の点 => 70 80 90
受験番号 => 102
英語　数学　国語の点 => 80 90 100
受験番号 => 103
英語　数学　国語の点 => 60 70 60
受験番号 => 0

```

- 作成された **seiseki.dat** の中身

```
101 70 80 90 240
102 80 90 100 270
103 60 70 60 190
```

````c
// e1-1.c 解答
#include <stdio.h>

int main(void)
{
    int    jno;
    int    ei, suu, kok, total;
    FILE   *fp;
    
    fp = fopen("seiseki.dat","w");
    printf("受験番号 => ");
    scanf("%d",&jno);
    
    while(jno>0) {
        printf("英語　数学　国語の点 => ");
        scanf("%d %d %d", &ei, &suu, &kok);
        total = ei + suu + kok;
        
        fprintf(fp,"%d %d %d %d %d\n", jno, ei, suu, kok, total);
        printf("受験番号 => ");
        scanf("%d", &jno);
    }
    
    fclose(fp);
    
    return 0;
}
````

## 

課題e1.1（前の問題）で作成した **seiseki.dat** を開き, ファイルから受験番号と英語, 数学, 国語の得点を読み出し，以下のように表示するプログラムを作成せよ．

実行例

```
受験番号　英語　数学　国語　合計
    101    70    80    90   240
    102    80    90   100   270
    103    60    70    60   190
```

````c
// e1-2.c 解答
#include <stdio.h>

int  main(void)
{
    int    jno;
    int    ei, suu, kok, kei;
    FILE   *fp;
    fp = fopen("seiseki.dat","r");
    
    printf("受験番号　英語　数学　国語　合計\n");
    while(fscanf(fp,"%d %d %d %d %d", &jno, &ei, &suu, &kok, &kei) != EOF)
    {
        kei = ei + suu + kok;
        printf("%7d %5d %5d %5d %5d\n", jno, ei, suu, kok, kei);
    }
    fclose(fp);
    return 0;
}
````

````c
// e1-7.c 解答
#include <stdio.h>

struct  member{
    int     no;          // 番号
    char    name[10];    // 氏名
    int     age;         // 年齢
    int     exp;         // 勤務年数
};

//　構造体の1件分のデータをファイルに書き込む関数
void disp(struct member *p);

char *fname = "meibo2.csv";

int main(void)
{
    int i;
    FILE *fp;
    struct  member p[5]={
        {1, "田中", 21, 4},
        {2, "山田", 19, 3},
        {3, "坂本", 20, 5},
        {4, "岡本", 30, 10},
        {5, "佐藤", 50, 18}
    };
    
    // 最初ファイルを書き込み専用で開き， 閉じることで中身をクリアする
    fp = fopen(fname,"w");
    fclose(fp);

    for(i=0;i<5;i++){
        /* 構造体のアドレス渡しを用いる*/
        disp(&p[i]);
    }
    
    return 0;
}

void disp(struct member *p)
{
    FILE *fp;
    
    /* 追加書き込みでファイルを開く*/
    fp = fopen(fname,"a");
    
    /* データ一件分を表示 */
    fprintf(fp, "%d, %s, %d, %d\n",p->no, p->name, p->age, p->exp);
    
    printf("1件分のデータを書き込み\n");
    fclose(fp);
}
````

````c
// e2-3.c 解答
#include <stdlib.h>
#include <stdio.h>
#include <math.h>

// 関数形式マクロの定義
// 0 〜 1のランダムな数値を返す
#define rand_value() (rand() / (RAND_MAX + 1.0))

//n回砂粒を落としてπを計算し返却する関数
double simulation(int n);

int main()
{
    int n;
    FILE *fp;
    char *fname = "result_e2-3.csv";
    
    double result, error;
    /* 乱数の種を初期化 */
    srand(1234);
    
    fp = fopen( fname, "w" );
    if( fp == NULL ){
        printf( "%sファイルが開けません\n", fname );
        return -1;
    }
    
    printf("n, pai, error\n");
    fprintf(fp, "n, pai, error\n");
    
    // いくつかのnでπの近似結果を表示
    for(n = 50; n<=1000; n +=50){
        result = simulation(n);
        error = fabs(M_PI - result);
        printf("%d, %f, %f\n", n, result, error);
        fprintf(fp,"%d, %f, %f\n", n, result,error);
        
    }
    fclose(fp);
    return 0;
}

double simulation(int n){
    int i, count=0;;
    double x, y, pi;
    
    // n回砂粒を落とす
    for (i=0; i<n; i++) {
        x = rand_value();   /* x座標を乱数で計算 */
        y = rand_value();   /* y座標を乱数で計算 */
        
        // 落とした砂粒が円の第１象限にある部分(図の斜線内)に入るか?
        if (y*y <= (1-x*x)){
            count++;          /* 斜線内ならカウント */
        }
    }
    pi = 4 * (count / (double)n); /* πの値を計算 */
    
    return pi;   /*  計算したπを返す*/
}
````

## 

以下は，区間$[𝑎,𝑏]$で連続な関数$𝑦=𝑓(𝑥)$の定積分を行うプログラムである．

課題eの説明のスライドを参考にし，プログラムを完成させよ．


```c
// サンプルコード
#include <stdlib.h>
#include <stdio.h>
#include <math.h>

// 対象の関数
double func(double x);

// 長方形近似による区分求積法
double rectangle(double a, double b, int n);

int main(void)
{
    int n;
    double a, b;
    double r1;
    
    // 区間[a,b]の面積を求める
    a = 0;
    b = 1;
    
    printf("n, rectangle\n");
    for(n=10; n<=100; n+=10){
        r1=rectangle(a, b, n);
        printf("%d, %f\n", n, r1);
    }
    
    return 0;
}

double func(double x){
    return(1 / (1 + x*x));
}

double rectangle(double a, double b, int n){
    double x, dx, h, sum, ds;
    
    dx = (b-a) / n;
    sum = 0;
    x = a;
    
    /* 面積の計算 */
    while (x < b) {
        h = func(x);
        ds = h * dx;
        sum += ds;
        x += dx;
    }
    return sum;
}
```

````c
// e2-4.c 解答
#include <stdlib.h>
#include <stdio.h>
#include <math.h>

// 対象の関数
double func(double x);

// 長方形近似による区分求積法
double rectangle(double a, double b, int n);

// 台形近似による区分求積法
double trapezoid(double a, double b, int n);


int main(void)
{
    int n;
    double a, b;
    double r1, r2, e1, e2;
    FILE *fp;
    char *fname = "result_e2-4.csv";
    
    fp = fopen( fname, "w" );
    if( fp == NULL ){
        printf( "%sファイルが開けません\n", fname );
        return -1;
    }

    // 区間[a,b]の面積を求める
    a = 0;
    b = 1;
    
    fprintf(fp, "n, rectangle, trapezoid, rectangle_error, trapezoid_error\n");
    printf("n, rectangle, trapezoid, rectangle_error, trapezoid_error\n");
    for(n=10; n<=100; n+=10){
        r1=rectangle(a, b, n);
        r2=trapezoid(a, b, n);
        e1=fabs(r1-M_PI/4);
        e2=fabs(r2-M_PI/4);
        printf("%d, %f, %f, %f, %f\n", n, r1,r2,e1,e2);
        fprintf(fp,"%d, %f, %f, %f, %f\n", n, r1,r2,e1,e2);
    }
    
    fclose(fp);
    return 0;
}

double func(double x){
    return(1 / (1 + x*x));
}

double rectangle(double a, double b, int n){
    double x, dx, h, sum, ds;
    
    dx = (b-a) / n;
    sum = 0;
    x = a;
    
    /* 面積の計算 */
    while (x < b) {
        h = func(x);
        ds = h * dx;
        sum += ds;
        x += dx;
    }
    return sum;
}

double trapezoid(double a, double b, int n){
    double x, dx, h1, h2, sum, ds;
    
    dx = (b-a) / n;
    sum = 0;
    x = a;
    
    /* 面積の計算 */
    while (x < b) {
        h1 = func(x);
        h2 = func(x + dx);
        ds = (h1 + h2) * dx / 2;
        sum += ds;
        x += dx;
    }
    
    return sum;
}
````
