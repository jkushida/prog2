````c
// e2-4.c è§£ç­”
#include <stdlib.h>
#include <stdio.h>
#include <math.h>

// å¯¾è±¡ã®é–¢æ•°
double func(double x);

// é•·æ–¹å½¢è¿‘ä¼¼ã«ã‚ˆã‚‹åŒºåˆ†æ±‚ç©æ³•
double rectangle(double a, double b, int n);

// å°å½¢è¿‘ä¼¼ã«ã‚ˆã‚‹åŒºåˆ†æ±‚ç©æ³•
double trapezoid(double a, double b, int n);


int main(void)
{
    int n;
    double a, b;
    double r1, r2, e1, e2;
    FILE *fp;
    char *fname = "result_e2-4.csv";
    
    fp = fopen( fname, "w" );
    if( fp == NULL ){
        printf( "%sãƒ•ã‚¡ã‚¤ãƒ«ãŒé–‹ã‘ã¾ã›ã‚“\n", fname );
        return -1;
    }

    // åŒºé–“[a,b]ã®é¢ç©ã‚’æ±‚ã‚ã‚‹
    a = 0;
    b = 1;
    
    fprintf(fp, "n, rectangle, trapezoid, rectangle_error, trapezoid_error\n");
    printf("n, rectangle, trapezoid, rectangle_error, trapezoid_error\n");
    for(n=10; n<=100; n+=10){
        r1=rectangle(a, b, n);
        r2=trapezoid(a, b, n);
        e1=fabs(r1-M_PI/4);
        e2=fabs(r2-M_PI/4);
        printf("%d, %f, %f, %f, %f\n", n, r1,r2,e1,e2);
        fprintf(fp,"%d, %f, %f, %f, %f\n", n, r1,r2,e1,e2);
    }
    
    fclose(fp);
    return 0;
}

double func(double x){
    return(1 / (1 + x*x));
}

double rectangle(double a, double b, int n){
    double x, dx, h, sum, ds;
    
    dx = (b-a) / n;
    sum = 0;
    x = a;
    
    /* é¢ç©ã®è¨ˆç®— */
    while (x < b) {
        h = func(x);
        ds = h * dx;
        sum += ds;
        x += dx;
    }
    return sum;
}

double trapezoid(double a, double b, int n){
    double x, dx, h1, h2, sum, ds;
    
    dx = (b-a) / n;
    sum = 0;
    x = a;
    
    /* é¢ç©ã®è¨ˆç®— */
    while (x < b) {
        h1 = func(x);
        h2 = func(x + dx);
        ds = (h1 + h2) * dx / 2;
        sum += ds;
        x += dx;
    }
    
    return sum;
}
````
