# ポインタ

## 

次のプログラムにコードを追加し, 9と表示されるプログラムを完成させよ．

ただし, 変数`x`に数値 9 を直接代入するのではなく, ポインタ変数`p`を使用して`x`の値を 9 にすること．

```c
#include <stdio.h>

int main(void)
{
    int x;
    int *p;

    // xの値を9にする

    printf( "%d\n", x );
    return 0;
}

```

````c
// c1-1.c 解答
#include <stdio.h>

int main(void)
{
    int x;
    int *p;

    // xの値を9にする
    p = &x;
    *p = 9;

    printf( "%d\n", x );
    return 0;
}
````

以下のプログラムでは，文字型配列`str1`に, "abcde" という文字列が格納されている．

ポインタを２つ用いて, 文字型配列`str2`に`str1`の要素を逆順に格納し，表示させよ．

```c
#include <stdio.h>

int main(void)
{
    char str1[] = "abcde";
    char str2[6] = {0};
    char *p1, *p2;

    /* コードを書く */
    
    printf("%s\n",p2);

    return 0;
}
```

実行結果
```c
edcba
```

````c
// c1-5.c 解答
#include <stdio.h>

int main(void)
{
    char str1[] = "abcde";
    char str2[6] = {0};
    char *p1, *p2;
    int i, len = 0;
    
    p2 = str2;
    for(p1 = str1; *p1 != '\0' ;p1++ )
        len++;
    p1--;
    
    for(i=0;i<len;i++){
        *p2 = *p1;
        p2++;
        p1--;
    }
    
    p2 = str2;
    
    printf("%s\n",p2);
    
    return 0;
}
````

## 

以下は`malloc`関数で, **int型の要素を格納するメモリ領域を動的に確保する**プログラムである．

実行結果と同様の出力が表示されるように, プログラムを完成させよ．

```c
#include <stdio.h>
#include <stdlib.h>

int main(void)
{
    int *ip;        // int型のポインタipを宣言
    int i, n;
    
    n = 5;
    ip = (int *)malloc(n * sizeof(int));  // メモリ領域の確保
    
    for (i=0; i<n; i++){
        /* コードを書く */
    }
    
    for (i=0; i<n; i++)
        printf("%d ", ip[i]);
    printf("\n");
    
    free(ip);       // 確保したメモリ領域の解放
    
    return 0;
}
```

実行結果
```
5 4 3 2 1 
```

````c
// c2-1.c 解答
#include <stdio.h>
#include <stdlib.h>

int main(void)
{
    int *ip;        // int型のポインタipを宣言
    int i, n;

    n = 5;
    ip = (int *)malloc(n * sizeof(int));  // メモリ領域の確保

    for (i=0; i<n; i++){
        ip[i] = n-i;
    }

    for (i=0; i<n; i++)
        printf("%d ", ip[i]);
    printf("\n");

    free(ip);       // 確保したメモリ領域の解放

    return 0;
}
````

````c
// c4-4.c 解答
#include <stdio.h>

void func(char *str);

int main(void)
{
    char str[ ] = "ABCD";
    func(str);
    printf("\n");
    return 0;
}

void func(char *str)
{
    char *head = str;
    while(*str != '\0'){
        str++;
    }
    while(str != head){
        str--;
        printf("%c", *str);
    }
}
````

````c
// c4-5.c 解答
#include <stdio.h>

#define N 5

void func(int *a1, int *a2, int *a3);

int main(void)
{
    int array1[N]={1, 2, 3, 4, 5};
    int array2[N]={2, 4, 6, 8, 10};
    int i, array3[N];

    func(array1, array2, array3);
    
    for(i=0; i<N; i++)
        printf("%d ", array3[i]);
    printf("\n");
    
    return 0;
}

void func(int *a1, int *a2, int *a3)
{
    int i;

    for(i=0; i<N; ++i)
        *(a3 + i) = *(a1 + i) + *(a2 + i);
}
````

````c
// c4-6.c 解答
#include <stdlib.h>
#include <stdio.h>

#define N 5

int *func();

int main(void)
{
    int i, *y;

    y = func(); // 関数からアドレスを受け取る
    
    for(i=0; i<N; i++ )
        printf("%d ", *(y + i));
    printf("\n");

    return 0;
}

// N個のint型変数のためのメモリ領域を確保し値を代入する関数
int *func()
{
    int i, *x;
    x = (int *)malloc(N * sizeof(int));  // メモリ領域の確保
    for(i=0; i<N; i++)
        x[i] = i;
        
    // xの先頭アドレスを返す
    return x;
}
````
